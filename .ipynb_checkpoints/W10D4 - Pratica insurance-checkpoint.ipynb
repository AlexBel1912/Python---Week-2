```python
"""Esercizio 
Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini delle persone, e della zona in cui vivono, rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute. 
• Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
• E rispetto a smoker? E a sex? 
• Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
"""
import os
import pandas as pd

# Carica il dataset
try:
    tabella = pd.read_csv(r"insurance.csv")
    print("Dataset caricato con successo.")
except FileNotFoundError:
    print("Errore: File insurance.csv non trovato. Assicurati che il file sia nella stessa directory dello script.")
    exit()
except Exception as e:
    print(f"Errore durante il caricamento del dataset: {e}")
    exit()

#Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
numero_righe, numero_colonne = tabella.shape
print(f"Il dataset contiene {numero_righe} righe e {numero_colonne} colonne." )
display(tabella)

print("\nNomi delle colonne:")
print(tabella.columns)
```

```python
#Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
medie_charges_region = tabella.groupby('region')['charges'].mean()

print("\nMedie di charges rispetto a region:")
print(medie_charges_region)

# Verifica differenze significative (esempio con deviazione standard)
deviazioni_charges_region = tabella.groupby('region')['charges'].std()
print("\nDeviazioni standard di charges rispetto a region:")
print(deviazioni_charges_region)
```

```python
#E rispetto a smoker? E a sex? 

# Medie di charges rispetto a smoker
medie_charges_smoker = tabella.groupby('smoker')['charges'].mean()
print("\nMedie di charges rispetto a smoker:")
print(medie_charges_smoker)

# Medie di charges rispetto a sex
medie_charges_sex = tabella.groupby('sex')['charges'].mean()
print("\nMedie di charges rispetto a sex:")
print(medie_charges_sex)
```

```python
# Quali sono i descrittori statistici di bmi? 
print("\nDescrittori statistici di bmi:")
print(tabella['bmi'].describe())

#Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
quartili_bmi = tabella['bmi'].quantile([0.25, 0.5, 0.75])
print("\nQuartili di bmi:")
print(quartili_bmi)

# Analisi di charges rispetto ai quartili di bmi
def analisi_charges_per_quartile(bmi):
    if bmi <= quartili_bmi[0.25]:
        return 'Q1'
    elif bmi <= quartili_bmi[0.5]:
        return 'Q2'
    elif bmi <= quartili_bmi[0.75]:
        return 'Q3'
    else:
        return 'Q4'
tabella['bmi_quartile'] = tabella['bmi'].apply(analisi_charges_per_quartile)

# Calcola le statistiche di charges per quartile di bmi
statistiche_charges_bmi = tabella.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max'])
print("\nStatistiche di charges rispetto ai quartili di bmi:")
print(statistiche_charges_bmi)
```
