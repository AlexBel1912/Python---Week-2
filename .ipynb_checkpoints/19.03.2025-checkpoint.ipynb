```python
import os
import dotenv
import sqlalchemy
import pandas as pd

#carico il file di configurazione
#dotenv.load_dotenv("config_env.txt")
dotenv.load_dotenv(override = True, dotenv_path = "config_env.txt") #caricamento del file .env

#recupero i dati al suo interno e li inserisco dentro le variabile nello script
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

print("username e uguale a:", username)

#costruire la conessione (stringa)
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"

try:
    #provo a creare un motore di connessione
    db_engine = sqlalchemy.create_engine(connection_string)
    #mi provo a collegare al database
    print("sono collegato al database!")
    query = "SELECT * FROM dimproduct"  # stringa che contiene un'interrogazione SQL
    tabella = pd.read_sql(query, db_engine) # utilizziamo pandas per leggere il result set del mio statement SQL
    print(tabella)
 

except Exception as e:
    # se non ci riesco lancio un eccezione 
    print("Errore di connessione al db")
    
```

```python
filtro_colore = tabella.loc[:, "Color"] == "Blue"
tabella_col_price = tabella.loc[filtro_colore, "DealerPrice"]
print("Prezzo medio dei prodotti di colore blu:", tabella_col_price.mean().round(2))
```

```python
#Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50; 
# Suddivisione in fasce di età
filtro_10 = diabetes["Age (years)"] < 20
filtro_20 = (20 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 29)
filtro_30 = (30 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 39)
filtro_40 = (40 <= diabetes["Age (years)"]) & (diabetes["Age (years)"] <= 49)
filtro_50 = diabetes["Age (years)"] >= 50
dpb10 = diabetes.loc[filtro_10, "Diastolic blood pressure (mm Hg)"].mean()
dpb20 = diabetes.loc[filtro_20, "Diastolic blood pressure (mm Hg)"].mean()
dpb30 = diabetes.loc[filtro_30, "Diastolic blood pressure (mm Hg)"].mean()
dpb40 = diabetes.loc[filtro_40, "Diastolic blood pressure (mm Hg)"].mean()
dpb50 = diabetes.loc[filtro_50, "Diastolic blood pressure (mm Hg)"].mean()
print(dpb10, dpb20, dpb30, dpb40, dpb50)
```

```python
filtro_sw = ins["region"] == "southwest"
filtro_se = ins["region"] == "southeast"  # Corretto
filtro_nw = ins["region"] == "northwest"
filtro_ne = ins["region"] == "northeast"  # Corretto

ch_sw = ins.loc[filtro_sw, "charges"].mean()
ch_se = ins.loc[filtro_se, "charges"].mean()
ch_nw = ins.loc[filtro_nw, "charges"].mean()
ch_ne = ins.loc[filtro_ne, "charges"].mean()

print("SW", ch_sw)
print("SE", ch_se)
print("NW", ch_nw)
print("NE", ch_ne)
```
