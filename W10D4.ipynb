```python
"""Esercizio 1/3 
Estraiamo la tabella dimproduct dal database AdventureWorks, e valutiamo quanto segue: 
• Quanti dati ci sono in totale? 
• Quali sono i metadati? 
• Stampiamo il primo elemento 
• Stampiamo l'ultimo elemento 
• Riusciamo a stampare cinque elementi a caso? 
• Quali sono i colori disponibili?"""

import os
import dotenv
import sqlalchemy
import pandas as pd

#carico il file di configurazione
#dotenv.load_dotenv("config_env.txt")
dotenv.load_dotenv(override = True, dotenv_path = "config_env.txt") #caricamento del file .env

#recupero i dati al suo interno e li inserisco dentro le variabile nello script
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

print("username e uguale a:", username)

#costruire la conessione (stringa)
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"

try:
    #provo a creare un motore di connessione
    db_engine = sqlalchemy.create_engine(connection_string)
    
    #mi provo a collegare al database
    print("sono collegato al database!")
    #estrai la tabella dimproduct in un Dataframe di pandas
    tabella = pd.read_sql("dimproduct", db_engine) 
    print(tabella)

    # 1. Quanti dati ci sono in totale? 
    numero_rows, numero_coloumns = tabella.shape
    print(f"Totale dati: {numero_rows} righe, {numero_coloumns} colonne")

    # 2. Quali sono i metadati?
    print("\nMetadati (colonne):")
    print(tabella.dtypes)
    
    # 3. Stampiamo il primo elemento 
    print("\nPrimo elemento:")
    print(tabella.iloc[0])

    # 4. Stampa l'ultimo elemento
    print("\nUltimo elemento:")
    print(tabella.iloc[-1])

    # 5. Riusciamo a stampare cinque elementi a caso? 
    print("\nCinque elementi a caso:")
    print(tabella.sample(5))

    # 6. Quali sono i colori disponibili?
    colori = tabella["Color"].unique()
    print("\nColori disponibili:")
    print(colori)

except Exception as e:
    # se non ci riesco lancio un eccezione 
    print("Errore di connessione al db")
```

```python
#Esercizio 2/3 
# In media quanto pesano i prodotti? 
peso_medio = tabella["Weight"].mean()
print(f"peso medio dei prodotti: {peso_medio:.2f}")

# Quanto pesa il più leggero? 
peso_minimo = tabella["Weight"].min()
print(f"peso minimo dei prodotti: {peso_minimo:.2f}")

# Quanto pesa il più pesante? 
peso_massimo = tabella["Weight"].max()
print(f"peso massimo dei prodotti: {peso_massimo:.2f}")

# Quanti prodotti pesano più di 100 Kg? 
prodotti_sopra_100kg = tabella[tabella["Weight"] > 100]
x = len(prodotti_sopra_100kg)
print(f"Ci sono {x} prodotti che pesano piu di 100 kg")

# Quanto costano in media i prodotti (colonna DealerPrice)? 
prezzo_medio = tabella["DealerPrice"].mean()
print(f"{prezzo_medio: .2f} rappresenta il costo medio dei prodotti")

# Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?
prodotti_ordinati_per_prezzo = tabella.sort_values(by="DealerPrice", ascending=False)
un_quarto=len(prodotti_ordinati_per_prezzo) // 4
prodotti_piu_costosi = prodotti_ordinati_per_prezzo.head(un_quarto)
prezzo_minimo = prodotti_piu_costosi["DealerPrice"].min()
prezzo_massimo = prodotti_piu_costosi["DealerPrice"].max()
print(f"Un quarto di tutti i prodotti, che rappresentano i piu costosi, hanno il prezzo compreso tra {prezzo_minimo: .2f} e {prezzo_massimo: .2f}")
```

```python
#Esercizio 3/3 
#• Qual è il prezzo medio per i prodotti di colore blu? 

prezzo_medio_prodotti_blu = tabella[tabella["Color"] == "Blue"]["ListPrice"].mean()
print(f"{prezzo_medio_prodotti_blu: .2f} e il prezzo medio dei prodotti blu")

#• Qual è il prezzo medio per i prodotti di colore rosso o nero? 

prezzo_medio_prodotti_rosso_nero = tabella[tabella["Color"].isin(["Red", "Black"])]["ListPrice"].mean()
print(f"{prezzo_medio_prodotti_rosso_nero: .2f} e il prezzo medio dei prodotti rosso o nero")

#• Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? 

prezzo_massimo = tabella[(tabella["Size"] == "42") & (tabella["Weight"] > 10)]["ListPrice"].max()
print(f"{prezzo_massimo} rappresenta il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 kg")

#• Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? 

prodotti = tabella[(tabella["Size"] == "42") & (tabella["Weight"] > 10) & (tabella["Color"] == "Silver")]
for index, row in prodotti.iterrows():
    print(f"Il nome inglese e {row['EnglishProductName']} ed il costo di produzione e {row['StandardCost']:.2f}")

#• Esaminiamo la differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: c'è un pattern? C'è qualche elemento che non lo segue?
ultimi_20 = tabella.tail(20)
differenze = ultimi_20["DealerPrice"] - ultimi_20["StandardCost"]
print("\nLa differenza tra lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset:")
print(differenze)
print(ultimi_20[['EnglishProductName', 'StandardCost', 'ListPrice', 'DealerPrice']], differenze)

"""
query = "SELECT DealerPrice, StandardCost FROM dimproduct"
tabella_differenze = pd.read_sql(query, db_engine)
print(tabella_differenze)
"""

print(f"\nMedia Differenza: {differenze.mean():.2f}")
print(f"Deviazione standard differenza: {differenze.std():.2f}")

""" Annalizzando i risultati vediamo che la media e di 22.64 che indica, in media, il DealerPrice è superiore allo StandardCost di circa 22.64 dollari. 
Questo suggerisce che l'azienda applica un margine di profitto sui prodotti.La deviazione standard di 20.10 è relativamente alta rispetto alla media. 
Questo significa che la differenza tra StandardCost e DealerPrice varia notevolmente tra i prodotti. 
Una deviazione standard elevata indica che non c'è un margine di profitto fisso e uniforme applicato a tutti i prodotti. 
Alcuni prodotti hanno un margine molto più alto, mentre altri ne hanno uno più basso. Con una deviazione standard così alta, è difficile individuare un pattern preciso. 
Il margine di profitto sembra essere variabile e non segue una regola fissa. 
Potrebbero esserci diversi fattori che influenzano il margine di profitto, come il tipo di prodotto, la domanda, la concorrenza, ecc.
La deviazione standard elevata suggerisce che ci sono sicuramente elementi che si discostano dalla media."""
```
