```python
import pandas as pd

#definisco la stringa URL dove risiede il CSV
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

#aggiungo i nomi delle colonne
column_names = ["sepal_lenght", "sepal_width", "petal_length", "petal_width", "species"]
iris = pd.read_csv(url, names=column_names) # iris e il dataframe

print(iris)
display(iris)

print(iris.shape) #restituisce numero di righe e colonne
print(iris.describe()) #restituisce informazioni statistiche di base
```

```python
print(iris["species"].value_counts()) #contato quante righe per ogni specie
```

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

#carico il file di configurazione
#dotenv.load_dotenv("config_env.txt")
dotenv.load_dotenv(override = True, dotenv_path = "config_env.txt") #caricamento del file .env

#recupero i dati al suo interno e li inserisco dentro le variabile nello script
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

print("username e uguale a:", username)

#costruire la conessione (stringa)
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"

try:
    #provo a creare un motore di connessione
    db_engine = sqlalchemy.create_engine(connection_string)
    #mi provo a collegare al database
    print("sono collegato al database!")
    query = "SELECT * FROM dimproduct"  # stringa che contiene un'interrogazione SQL
    tabella = pd.read_sql(query, db_engine) # utilizziamo pandas per leggere il result set del mio statement SQL
    print(tabella)
    display (tabella)

    query_customer = "SELECT * FROM dimcustomer"
    customer_df = pd.read_sql(query_customer, db_engine)

    display(customer_df.head())
    print(customer_df.shape)
    df_condizione = customer_df.loc[customer_df["MaritalStatus"] == "M"]
    display(df_condizione)

    #casa de proprieta
    df_casa_proprieta = customer_df.loc[customer_df["HouseOwnerFlag"] == "1"]
    display(df_casa_proprieta)


except Exception as e:
    # se non ci riesco lancio un eccezione 
    print("Errore di connessione al db")
    
```
