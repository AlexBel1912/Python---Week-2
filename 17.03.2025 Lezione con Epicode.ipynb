```python
import pandas as pd

#definisco la stringa URL dove risiede il CSV
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

#aggiungo i nomi delle colonne
column_names = ["sepal_lenght", "sepal_width", "petal_length", "petal_width", "species"]
iris = pd.read_csv(url, names=column_names) # iris e il dataframe

print(iris)
display(iris)

print(iris.shape) #restituisce numero di righe e colonne
print(iris.describe()) #restituisce informazioni statistiche di base
```

```python
print(iris["species"].value_counts()) #contato quante ighe per ogni specie
```
