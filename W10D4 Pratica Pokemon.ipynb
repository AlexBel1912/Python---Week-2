```python
"""Esercizio 1/2 
Il dataset pokemon.csv contiene un database di Pokémon, con dati quali nome, tipi di attacco, valori di attacco/difesa/eccetera, e se sono o meno leggendari. 
• Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna; 
• È verosimile che la prima colonna dovrebbe essere un indice? 
• Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? 
• Se no, settare la prima colonna come indice.
"""
import os
import pandas as pd

# Carica il dataset
try:
    tabella = pd.read_csv(r"pokemon.csv")
    print("Dataset caricato con successo.")
except FileNotFoundError:
    print("Errore: File pokemon.csv non trovato. Assicurati che il file sia nella stessa directory dello script.")
    exit()
except Exception as e:
    print(f"Errore durante il caricamento del dataset: {e}")
    exit()
```

```python
#Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna; 
numero_righe, numero_colonne = tabella.shape
print(f"Il dataset contiene {numero_righe} righe e {numero_colonne} colonne." )
display(tabella)

print("\nNomi delle colonne:")
print(tabella.columns)


print(f"Dimensioni del dataset: {tabella.shape}")
print("\nPrime 5 righe del dataset:")
print(tabella.head())
```

```python
#È verosimile che la prima colonna dovrebbe essere un indice? Alla prima vista si pero, dopo possiamo osservare che ci sono numeri che si ripetono ma le righe non sono con dati uguali.
```

```python
#Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano? 
print("\nConfronto tra l'indice di Pandas e la prima colonna:")
print(tabella.iloc[:, 0].head())  # Stampiamo i primi valori della prima colonna
print(tabella.index[:5])  # Stampiamo i primi valori dell'indice di Pandas
```

```python
# Se no, settare la prima colonna come indice.
prima_colonna = tabella.columns[0]
if not tabella.iloc[:, 0].equals(pd.Series(tabella.index)):
    print("\nLa prima colonna non combacia con l'indice automatico di Pandas. La impostiamo come indice.")
    tabella.set_index(prima_colonna, inplace=True)
    print("\nNuova anteprima con la prima colonna come indice:")
    print(tabella.head())
else:
    print("\nLa prima colonna combacia già con l'indice di Pandas. Nessuna modifica necessaria.")
```

```python
"""Esercizio 2/2 
• Quali sono i Pokémon leggendari? 
• E quali sono i leggendari di tipo 1 Grass? 
• E leggendari di tipo 1 Ice o Fire? 
• Ordiniamo il dataset per la colonna Name; 
• Trasformiamo Name nell'indice; 
• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
"""

#Quali sono i Pokémon leggendari?
leggendari = tabella[tabella['Legendary'] == True]
print("I Pokémon leggendari sono:")
print(leggendari)
```

```python
#E quali sono i leggendari di tipo 1 Grass? 
leggendari_grass = leggendari[leggendari['Type 1'] == 'Grass']
print("\nPokémon leggendari di tipo 1 Grass:")
print(leggendari_grass)
```

```python
# E leggendari di tipo 1 Ice o Fire? 
leggendari_ice_fire = leggendari[(leggendari['Type 1'] == 'Ice') | (leggendari['Type 1'] == 'Fire')]
print("\nPokémon leggendari di tipo 1 Ice o Fire:")
print(leggendari_ice_fire)
```

```python
# Ordiniamo il dataset per la colonna Name; 
tabella.sort_values(by='Name', inplace=True)
print("\nDataset ordinato per nome:")
print(tabella)
```

```python
# Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?
pokemon_generazione1 = tabella[(tabella['Generation'] == 1) & (tabella['Attack'] > 50) & (tabella['HP'] < 60)]
print("\nPokémon della prima generazione con attacco > 50 e HP < 60:")
print(pokemon_generazione1)
```
